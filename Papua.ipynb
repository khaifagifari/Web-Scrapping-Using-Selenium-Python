```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait

DRIVER_PATH = '/usr/lib/chromium-browser/chromedriver'
driver = webdriver.Chrome(executable_path= DRIVER_PATH)
source= 'https://covid19.papua.go.id/'
driver.get(source)

element = WebDriverWait(driver, 20).until(
    EC.element_to_be_clickable((By.XPATH, "//*[@id='example2_length']/label/select/option[4]")));
element.click()
```

```python
# element.page_source

content = driver.page_source
soup = BeautifulSoup(content)
soup
div1 = soup.find('table', id='table')
div2 = soup.find('table', id='example2')

links1 = div1.find_all('td')
links2 = div2.find_all('td')

# soup
```

```python
date = soup.find('h4', class_ = 'float-right')
tempDate = date.get_text()
if 'Jul' in tempDate:
    tempDate = tempDate.replace(" Jul 2020","/07/2020")
    
from datetime import datetime
now = datetime.now().strftime('%d/%m/%Y')
# tempDate
```

```python
for i, link in enumerate(links1):    
    if (link.get_text()[:3] == 'Kab' or link.get_text()[:3] == 'Kot' or link.get_text()[:3] == 'Kep'):
        kota = link.get_text().replace('Kab. ','')
        idxKabkot = (kabkot.index(kota))
    if i in range (3, len(links1), 8):
        positif = re.findall(r'[0-9]+', link.get_text())
        total_positif[idxKabkot] = int(positif[0])
    if i in range (4, len(links1), 8):
        rawat = re.findall(r'[0-9]+', link.get_text())
        positif_dirawat[idxKabkot] = int(rawat[0])
    if i in range (5, len(links1), 8):
        sembuh = re.findall(r'[0-9]+', link.get_text())
        positif_sembuh[idxKabkot] = int(sembuh[0])
    if i in range (6, len(links1), 8):
        meninggal = re.findall(r'[0-9]+', link.get_text())
        positif_meninggal[idxKabkot] = int(meninggal[0])
    if i in range (7, len(links1), 8):
        pdp = re.findall(r'[0-9]+', link.get_text())
        total_pdp[idxKabkot] = int(pdp[0])
```
